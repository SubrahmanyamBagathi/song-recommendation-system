```python
import pandas as pd
df = pd.read_csv('spotify dataset.csv')
```

## Handling missing values

```python
print(df.isnull().sum())
df = df.dropna()
df = df.drop_duplicates()
df['track_album_release_date'] = pd.to_datetime(df['track_album_release_date'],format='mixed')
```

## scaling numerical features

```python
df['release_year'] = df['track_album_release_date'].dt.year
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
numerical_features = ['danceability', 'energy', 'loudness', 'speechiness', 'acousticness', 'instrumentalness', 'liveness', 'valence', 'tempo', 'duration_ms']
df[numerical_features] = scaler.fit_transform(df[numerical_features])
```

## data visualization

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
sns.histplot(df['track_popularity'], bins=30, kde=True)
plt.title('Distribution of Track Popularity')
plt.xlabel('Popularity')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.heatmap(df[numerical_features].corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Matrix of Numerical Features')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.lineplot(x='release_year', y='track_popularity', data=df)
plt.title('Track Popularity Over the Years')
plt.xlabel('Release Year')
plt.ylabel('Popularity')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(y='playlist_genre', data=df, order=df['playlist_genre'].value_counts().index)
plt.title('Distribution of Tracks Across Playlist Genres')
plt.xlabel('Count')
plt.ylabel('Playlist Genre')
plt.show()
```

## clustering

```python
from sklearn.cluster import KMeans

X = df[numerical_features]

inertia = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X)
    inertia.append(kmeans.inertia_)

plt.figure(figsize=(10, 6))
plt.plot(range(1, 11), inertia, marker='o')
plt.title('Elbow Method for Optimal k')
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia')
plt.show()

```

```python
kmeans = KMeans(n_clusters=4, random_state=42)
df['cluster'] = kmeans.fit_predict(X)

from sklearn.decomposition import PCA

pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

plt.figure(figsize=(10, 6))
sns.scatterplot(x=X_pca[:, 0], y=X_pca[:, 1], hue=df['cluster'], palette='viridis')
plt.title('Clusters of Tracks')
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')
plt.show()
```

## Model building for recomendation system

```python
from sklearn.metrics.pairwise import cosine_similarity

cosine_sim = cosine_similarity(df[numerical_features], df[numerical_features])

def get_recommendations(track_id, cosine_sim=cosine_sim):
    idx = df.index[df['track_id'] == track_id].tolist()[0]
    sim_scores = list(enumerate(cosine_sim[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    sim_scores = sim_scores[1:11]  
    track_indices = [i[0] for i in sim_scores]
    return df['track_name'].iloc[track_indices]
    
recommendations = get_recommendations('6f807x0ima9a1j3VPbc7VN')
print(recommendations)
```
